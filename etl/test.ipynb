```python
from pyspark.sql import SparkSession
from MySql import MySql
import os

MYSQL_JAR = os.path.abspath("../driver/mysql-connector-j-8.0.33.jar")

spark = (
    SparkSession.builder
    .config("spark.driver.extraClassPath", MYSQL_JAR)
    .config("spark.executor.extraClassPath", MYSQL_JAR)
    .getOrCreate()
)

mysql = MySql(spark)
data = [{"name": "son", "age": 18}, {"name": "linh", "age": 20}]
df = spark.createDataFrame(data)   # chuyển list dict -> DataFrame
mysql.insert("test", df)
```
