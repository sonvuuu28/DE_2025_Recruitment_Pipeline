```python
import uuid
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
from Cassandra import Cassandra

spark = (
    SparkSession.builder.config(
        "spark.jars.packages", "com.datastax.spark:spark-cassandra-connector_2.12:3.1.0"
    )
    .getOrCreate()
)

# Dữ liệu cơ bản
data = [
    {
        "bid": "1",
        "bn": "Chrome 103",
        "campaign_id": "24",
        "cd": "click",
        "custom_track": "click",
        "de": "UTF-8",
        "dl": "http://example.com/page1",
        "dt": "CandidatePortal",
        "ed": "",
        "ev": "",
        "group_id": "groupA",
        "id": "id1",
        "job_id": "101",
        "md": "",
        "publisher_id": "0",
        "rl": "",
        "sr": "",
        "ts": "2025-01-01 10:00:00",
        "tz": "+0700",
        "ua": "User Agent Here",
        "uid": "u1",
        "utm_campaign": "c1",
        "utm_content": "ct1",
        "utm_medium": "m1",
        "utm_source": "s1",
        "utm_term": "t1",
        "v": "1",
        "vp": "1920x1080"
    }
]

# Tạo DataFrame từ dữ liệu cơ bản
df = spark.createDataFrame(data)

# UDF để tạo UUID mới cho create_time
uuid_udf = udf(lambda: str(uuid.uuid1()), StringType())

# Thêm cột create_time mới mỗi lần insert
df = df.withColumn("create_time", uuid_udf())
```

```python
import time

cass = Cassandra(spark)
while True:
    cass.insert_random("tracking", df)
    time.sleep(5)
    print("Thêm thành công")
```
